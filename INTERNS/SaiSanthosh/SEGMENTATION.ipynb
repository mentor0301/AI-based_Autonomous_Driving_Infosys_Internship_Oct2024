```python
import torch
import torchvision
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
import glob
import os

# Load the pre-trained model
model = torchvision.models.segmentation.deeplabv3_resnet101(pretrained=True).eval()
# Define image directory
image_direc = '/content/drive/MyDrive/leftImg8bit_trainvaltest/leftImg8bit/val/frankfurt'
image_paths = glob.glob(os.path.join(image_direc, "*.png"))

# Loop through a subset of images
for i, path in enumerate(image_paths[:10]):
    # Open and transform the image
    image = Image.open(path)
    input_tensor = transform(image).unsqueeze(0)

    # Perform inference
    with torch.no_grad():
        output = model(input_tensor)['out'][0]

    # Get class predictions
    output_predictions = output.argmax(0).numpy()

    # Display original image
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title('Original Image')

    # Display segmentation output
    plt.subplot(1, 2, 2)
    plt.imshow(output_predictions, cmap='jet')
    plt.title('Segmentation Output')
    plt.show()
```
